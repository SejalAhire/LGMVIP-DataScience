# Importing Modules

```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

# Loading  Original Image

```python
img = cv2.imread("original.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(13,13))
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

# Converting Image to GrayScale

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(13,13))
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

# Inverting the Image

```python
#when a image is inverted it helps in studying the image with more precision and detail.
#The more detailed study is important for the sketch generation as the sketches need to be accurate and good. 
```

```python
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(13,13))
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```
